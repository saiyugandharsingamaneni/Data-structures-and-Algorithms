#### Adding keys and values to hash_table

```python
class hash_table:
    
    def __init__(self):
        self.MAX=200
        self.arr=[None for i in range(self.MAX)]
        
    def get_hash(self,key):
        h=0
        for char in key:
            h += ord(char)
        return h % self.MAX
    
    def add(self,key,value):
        h=self.get_hash(key)
        self.arr[h]=value
    
    def get(self,key):
        h=self.get_hash(key)
        return self.arr[h]
```

```python
t=hash_table()
t.add('march 6',130)
```

```python
t.get("march 6")
```

#### Adding and getting values in to hash_table and look at status of array

```python
class hash_table:
    
    def __init__(self):
        self.MAX=200
        self.arr=[None for i in range(self.MAX)]
        
    def get_hash(self,key):
        h=0
        for char in key:
            h += ord(char)
        return h % self.MAX
    
    def __setitem__(self,key,value):
        h=self.get_hash(key)
        self.arr[h]=value
    
    def __getitem__(self,key):
        h=self.get_hash(key)
        return self.arr[h]
```

```python
t=hash_table()
t['march 6']=130
t["march 7"]=123
t["march 9"]=153
t["dec 6"]=110
```

#### deleting item from hash_table

```python
class hash_table:
    
    def __init__(self):
        self.MAX=200
        self.arr=[None for i in range(self.MAX)]
        
    def get_hash(self,key):
        h=0
        for char in key:
            h += ord(char)
        return h % self.MAX
    
    def __setitem__(self,key,value):
        h=self.get_hash(key)
        self.arr[h]=value
    
    def __getitem__(self,key):
        h=self.get_hash(key)
        return self.arr[h]
    
    def __delitem__(self,key):
        h=self.get_hash(key)
        self.arr[h]=None
```

```python
t=hash_table()
t['march 6']=130
t["march 7"]=123
t["march 9"]=153
t["dec 6"]=110
```

```python
del t['march 6']
```
